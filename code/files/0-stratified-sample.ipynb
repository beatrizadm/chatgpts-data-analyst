# Stratified sample

Generating a stratified sample so that the data is 10 MB.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/elasComputacao/raio-x-dados/main/data/dados-brutos/historico_alunos_raiox.csv")
df
```

```python
df['nome_setor'].value_counts()
```

```python
# removing the sector that only has 1 occurrence
df_sample = df.query("nome_setor not in ['UNID. ACAD. DE ENGENHARIA DE PETRÓLEO']")

# Suppose you want to stratify based on 'nome_setor', ensuring the sample represents all industries
# First, you need to ensure that 'nome_setor' is not null to avoid errors
df_sample = df_sample.dropna(subset=['nome_setor'])

# Splitting the data (here, just simulating to obtain the stratified sample, without interest in the training/testing itself)
# 'test_size=0.2' means that the stratified sample will be 20% of the total data
_, stratified_sample = train_test_split(df_sample, test_size=0.2, stratify=df_sample['nome_setor'])
```

```python
stratified_sample['nome_setor'].value_counts()
```

```python
stratified_sample.reset_index(drop=True)
```

```python
stratified_sample.to_csv('historico_estratificado.csv', index=False)
```
